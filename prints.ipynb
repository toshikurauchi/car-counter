```python
import cv2
import os
```

```python
def save_video_frames(video_path, output_dir, num_frames, crop_area=None):
    # Abre o vídeo
    cap = cv2.VideoCapture(video_path)
    
    # Verifica se o vídeo foi aberto corretamente
    if not cap.isOpened():
        print("Erro ao abrir o vídeo")
        return

    # Obtém o número total de frames e a taxa de frames por segundo (FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Calcula os intervalos para capturar frames distribuídos uniformemente
    frame_positions = [int(total_frames * i / num_frames) for i in range(1, num_frames + 1)]

    # Cria o diretório de saída, se não existir
    os.makedirs(output_dir, exist_ok=True)

    frame_count = 0
    saved_count = 0

    while cap.isOpened() and saved_count < num_frames:
        ret, frame = cap.read()

        try:
            print(frame.shape)
        except:
            pass

        # Sai do loop se o vídeo terminou
        if not ret:
            break
        
        # Se o frame atual for um dos escolhidos para capturar
        if frame_count in frame_positions:
            # Aplica o crop se houver uma área definida
            if crop_area:
                x, y, w, h = crop_area
                frame = frame[y:y+h, x:x+w]

            # Salva a imagem no diretório de saída
            output_filename = os.path.join(output_dir, f"frame_{saved_count+1}.png")
            cv2.imwrite(output_filename, frame)
            saved_count += 1
        
        frame_count += 1

    cap.release()
    print(f"{saved_count} frames foram salvos em {output_dir}")
```

```python
video_path = "data/Lukka.mp4"
output_dir = "frames"
num_frames = 401
crop_area = (100, 100, 350, 600)

save_video_frames(video_path, output_dir, num_frames, crop_area)
```
